```python
import pandas as pd
```

```python
# Bring all the CSVs from the www.football-data.co.uk website, which includes the betting odds previous to each match.

aa = pd.read_csv("../tables/football-data.co.uk/17-18.csv")
bb = pd.read_csv("../tables/football-data.co.uk/18-19.csv")
cc = pd.read_csv("../tables/football-data.co.uk/19-20.csv")
dd = pd.read_csv("../tables/football-data.co.uk/20-21.csv")
ee = pd.read_csv("../tables/football-data.co.uk/21-22.csv")
ff = pd.read_csv("../tables/football-data.co.uk/22-23.csv")
gg = pd.read_csv("../tables/football-data.co.uk/16-17.csv")
hh = pd.read_csv("../tables/football-data.co.uk/15-16.csv")
ii = pd.read_csv("../tables/football-data.co.uk/14-15.csv")
jj = pd.read_csv("../tables/football-data.co.uk/13-14.csv")
kk = pd.read_csv("../tables/football-data.co.uk/12-13.csv")
ll = pd.read_csv("../tables/football-data.co.uk/11-12.csv")
mm = pd.read_csv("../tables/football-data.co.uk/10-11.csv")
nn = pd.read_csv("../tables/football-data.co.uk/23-24.csv")


# Concatenate all these sources into one dataframe named old.

old = pd.concat([mm, ll, kk, jj, ii, hh, gg, aa, bb, cc, dd, ee, ff, nn])

# See amounts of NaN values.

x = ["B365H", "B365D", "B365A", "BWH", "BWD", "BWA", "IWH", "IWD", "IWA", "WHH", "WHD", "WHA", "VCH", "VCD", "VCA"]

bettinghouses = old[x]

bettinghouses.isnull().sum().sort_values(ascending=False)
```

```python
# Drop columns "LBH", "LBD", and "LBA" as values under such columns are mostly NaN.

old.drop(["PSH", "PSD", "PSA"], axis=1, inplace=True)
```

```python
# The 'new' dataframe is the table created in the matchscore workbook where it contains all the data of each team's performance previous to the match.

new = pd.read_csv("../coding-csv/in_game_match_data.csv", index_col=[0])
new.replace({'Almeria': "Almería", 'Sociedad': "Real Sociedad", "Ath Bilbao":"Athletic Club", "Celta": "Celta Vigo", "Cadiz":"Cádiz", "Ath Madrid":"Atlético Madrid", "Betis":"Real Betis", "Alaves":"Alavés", "Vallecano":"Rayo Vallecano", "La Coruna":"Deportivo La Coruña", "Malaga":"Málaga", "Espanol":"Espanyol", "Leganes":"Leganés"}, inplace=True)


# In order to merge both dataframes, unique ID is created to be the matching column between both datasets.

new["ID"] = new['date'].astype(str) + '-' + new['home_team'] + '-' + new['away_team']
new = pd.concat([new["ID"], new.drop("ID", axis=1)], axis=1)
new
```

```python
# Only columns without NaN values will be used from the 'old' dataframe, to avoid the need to create 'false' data to replace them.

a = old[old.columns[1:4]]
b = old[x]


# This selection of columns will be called 'newold'.

newold = pd.concat([a, b], axis=1)
newold.reset_index(drop=True, inplace=True)
newold
```

```python
# Team names are differently written in both datasets, a change must be done to make them be equal (for merging purposes only).

newold.replace({'Almeria': "Almería", 'Sociedad': "Real Sociedad", "Ath Bilbao":"Athletic Club", "Celta": "Celta Vigo", "Cadiz":"Cádiz", "Ath Madrid":"Atlético Madrid", "Betis":"Real Betis", "Alaves":"Alavés", "Vallecano":"Rayo Vallecano", "La Coruna":"Deportivo La Coruña", "Malaga":"Málaga", "Espanol":"Espanyol", "Leganes":"Leganés"}, inplace=True)


# Change the 'date' column type.

newold["Date"] = pd.to_datetime(newold["Date"], format='%d/%m/%Y', errors='coerce').fillna(pd.to_datetime(newold["Date"], format='%d/%m/%y', errors='coerce'))


# Create the ID column in the second dataframe.

newold["ID"] = newold['Date'].astype(str) + '-' + newold['HomeTeam'] + '-' + newold['AwayTeam']
newold = pd.concat([newold["ID"], newold.drop("ID", axis=1)], axis=1)
newold
```

```python
# Perform the merge between both datasets.

mergeddatasets = pd.merge(new, newold, on="ID", how='left')


# Drop columns that are duplicate from the 'newold' dataset.

mergeddatasets.drop(["Date", "HomeTeam", "AwayTeam"], axis=1, inplace=True)
```

```python
mergeddatasets.to_csv("../coding-csv/merged_datasets.csv")
```
