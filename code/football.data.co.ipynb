```python
thedata.sort_index()
import numpy as np

thedata['result'] = np.where(thedata['score'] > thedata['score_against'], 1, np.where(thedata['score'] == thedata['score_against'], 0.5, 0))
result = thedata['result']
thedata = thedata.drop(["result"], axis=1)
```

```python
def olddata (team):
    
    a = thedata[thedata["team"]==team]
    a['goals_offense'] = round(((a['score'].rolling(window=10, min_periods=0).mean()*0.6) + (a['score'].rolling(window=5, min_periods=0).mean()*0.3) + (a['score'].rolling(window=3, min_periods=0).mean()*0.1)), 2)
    a['goals_defense'] = round(((a['score_against'].rolling(window=10, min_periods=0).mean()*0.6) + (a['score_against'].rolling(window=5, min_periods=0).mean()*0.3) + (a['score_against'].rolling(window=3, min_periods=0).mean()*0.1)), 1)

    new = a[["Date", "team", "goals_offense", "goals_defense"]]

    b = thedata[thedata.columns[3:]]
    c = thedata[thedata.columns[:3]]

    for i in b.columns:
        b[i] = round(b[i] / ((b[i].rolling(window=100, min_periods=0).quantile(0.95)*0.6) + (b[i].rolling(window=50, min_periods=0).quantile(0.95)*0.3) + (b[i].rolling(window=30, min_periods=0).quantile(0.95)*0.1)), 2)

    b = b.fillna(0)
    all = pd.concat([c, b], axis=1)
    
    all['fouls'] = all['fouls'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['opponent_shots'] = all['opponent_shots'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['opponent_shots_on_target'] = all['opponent_shots_on_target'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['reds'] = all['reds'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['yellows'] = all['yellows'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))

    all["offense"] = all[["shots", "shots_on_target", "corners"]].mean(axis=1).round(2)
    all["in_game"] = all[["fouls", "reds", "yellows"]].mean(axis=1).round(2)
    all["defense"] = all[["opponent_shots", "opponent_shots_on_target"]].mean(axis=1).round(2)

    all = all[["team", "score", "score_against", "offense", "in_game", "defense"]]

    all = all[all["team"]==team]

    all["offense"] = round(((all['offense'].rolling(window=100, min_periods=0).mean()*0.6) + (all['offense'].rolling(window=50, min_periods=0).mean()*0.3) + (all['offense'].rolling(window=30, min_periods=0).mean()*0.1))*(1 + (all["score"] * 0.075)), 2)
    all["in_game"] = round(((all['in_game'].rolling(window=100, min_periods=0).mean()*0.6) + (all['in_game'].rolling(window=50, min_periods=0).mean()*0.3) + (all['in_game'].rolling(window=30, min_periods=0).mean()*0.1)), 2)
    all["defense"] = round(((all['defense'].rolling(window=100, min_periods=0).mean()*0.6) + (all['defense'].rolling(window=50, min_periods=0).mean()*0.3) + (all['defense'].rolling(window=30, min_periods=0).mean()*0.1))*(1 - (all["score_against"] * 0.075)), 2)

    all.drop(["team", "score", "score_against"], axis=1, inplace=True)

    complete = pd.concat([new, all], axis=1)

    return complete
```

```python
teams = thedata["team"].unique()

concatenate = []

for i in teams:
    if len(concatenate) == 0:
        concatenate = olddata(i)
    else:
        concatenate = pd.concat([concatenate, olddata(i)])
```

```python
concatenate.sort_index(inplace=True)
concatenate["result"] = result
concatenate
```

```python
df_even = concatenate[concatenate.index % 2 == 0]
df_odd = concatenate[concatenate.index % 2 == 1]

df_even.reset_index(drop=True, inplace=True)
df_odd.reset_index(drop=True, inplace=True)
```

```python
result = pd.merge(df_even, df_odd, left_index=True, right_index=True)

result.rename(columns={"Date_x":"date", "team_x":"home_team", "goals_offense_x":"home_goals_offense", "goals_defense_x":"home_goals_defense", "offense_x":"home_offense", "in_game_x":"home_in_game", "defense_x":"home_defense", "team_y":"away_team", "goals_offense_y":"away_goals_offense", "goals_defense_y":"away_goals_defense", "offense_y":"away_offense", "in_game_y":"away_in_game", "defense_y":"away_defense", "result_x":"result"}, inplace=True)
result.drop(["Date_y", "result_y"], axis=1, inplace=True)
result = pd.concat([result.drop("result", axis=1), result["result"]], axis=1)
result
```
