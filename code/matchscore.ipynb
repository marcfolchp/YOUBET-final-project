```python
def ratings (team):

    # Import the tables that were extracted from the scraping function notebook.

    aa = pd.read_csv("../tables/fbref.com/17-18.csv", index_col=[0]).reset_index().drop("index", axis=1)
    bb = pd.read_csv("../tables/fbref.com/18-19.csv", index_col=[0]).reset_index().drop("index", axis=1)
    cc = pd.read_csv("../tables/fbref.com/19-20.csv", index_col=[0]).reset_index().drop("index", axis=1)
    dd = pd.read_csv("../tables/fbref.com/20-21.csv", index_col=[0]).reset_index().drop("index", axis=1)
    ee = pd.read_csv("../tables/fbref.com/21-22.csv", index_col=[0]).reset_index().drop("index", axis=1)
    ff = pd.read_csv("../tables/fbref.com/22-23.csv", index_col=[0]).reset_index().drop("index", axis=1)
    gg = pd.read_csv("../tables/fbref.com/23-24.csv", index_col=[0]).reset_index().drop("index", axis=1)

    data = pd.concat([aa, bb, cc, dd, ee, ff, gg])

    new_row_data = {'team': 'Getafe', 'matchday': 16, 'date': '2023-12-08', "location":"Home", "opponent":"Valencia"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Valencia', 'matchday': 16, 'date': '2023-12-08', "location":"Away", "opponent":"Getafe"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Alavés', 'matchday': 16, 'date': '2023-12-09', "location":"Home", "opponent":"Las Palmas"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Las Palmas', 'matchday': 16, 'date': '2023-12-09', "location":"Away", "opponent":"Alavés"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Real Betis', 'matchday': 16, 'date': '2023-12-09', "location":"Home", "opponent":"Real Madrid"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Real Madrid', 'matchday': 16, 'date': '2023-12-09', "location":"Away", "opponent":"Real Betis"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Villarreal', 'matchday': 16, 'date': '2023-12-09', "location":"Home", "opponent":"Real Sociedad"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Real Sociedad', 'matchday': 16, 'date': '2023-12-09', "location":"Away", "opponent":"Villarreal"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Mallorca', 'matchday': 16, 'date': '2023-12-09', "location":"Home", "opponent":"Sevilla"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Sevilla', 'matchday': 16, 'date': '2023-12-09', "location":"Away", "opponent":"Mallorca"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Atlético Madrid', 'matchday': 16, 'date': '2023-12-10', "location":"Home", "opponent":"Almería"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Almería', 'matchday': 16, 'date': '2023-12-10', "location":"Away", "opponent":"Atlético Madrid"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Granada', 'matchday': 16, 'date': '2023-12-10', "location":"Home", "opponent":"Athletic Club"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Athletic Club', 'matchday': 16, 'date': '2023-12-10', "location":"Away", "opponent":"Granada"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Cádiz', 'matchday': 16, 'date': '2023-12-10', "location":"Home", "opponent":"Osasuna"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Osasuna', 'matchday': 16, 'date': '2023-12-10', "location":"Away", "opponent":"Cádiz"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Barcelona', 'matchday': 16, 'date': '2023-12-10', "location":"Home", "opponent":"Girona"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Girona', 'matchday': 16, 'date': '2023-12-10', "location":"Away", "opponent":"Barcelona"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Rayo Vallecano', 'matchday': 16, 'date': '2023-12-11', "location":"Home", "opponent":"Celta Vigo"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    new_row_data = {'team': 'Celta Vigo', 'matchday': 16, 'date': '2023-12-11', "location":"Away", "opponent":"Rayo Vallecano"}
    new_row_df = pd.DataFrame(new_row_data, index=[0])
    data = pd.concat([data, new_row_df], ignore_index=True)

    # Divide the concatenation by teams, to be able to calculate the averages, per team, of each match statistic brought from the scraping. 
    # This is done because we want to know a team's form coming into a match, that is, whether they have been playing good or bad lately.

    a = data[data["team"]==team]
    a['goals_offense'] = round((((a['score'].shift(1).rolling(window=10, min_periods=1).mean()*0.5) + (a['score'].shift(1).rolling(window=5, min_periods=1).mean()*0.4) + (a['score'].shift(1).rolling(window=3, min_periods=1).mean()*0.1))*0.7) + (((a['xG'].shift(1).rolling(window=10, min_periods=1).mean()*0.6) + (a['xG'].shift(1).rolling(window=5, min_periods=1).mean()*0.3) + (a['xG'].shift(1).rolling(window=3, min_periods=1).mean()*0.1))*0.3), 1)
    a['goals_defense'] = round((((a['score_against'].shift(1).rolling(window=10, min_periods=1).mean()*0.5) + (a['score_against'].shift(1).rolling(window=5, min_periods=1).mean()*0.4) + (a['score_against'].shift(1).rolling(window=3, min_periods=1).mean()*0.1))*0.7) + (((a['xg_against'].shift(1).rolling(window=10, min_periods=1).mean()*0.6) + (a['xg_against'].shift(1).rolling(window=5, min_periods=1).mean()*0.3) + (a['xg_against'].shift(1).rolling(window=3, min_periods=1).mean()*0.1))*0.3), 1)

    a["PSxG_2"] = round(((a['PSxG'].shift(1).rolling(window=100, min_periods=1).mean()*0.6) + (a['PSxG'].shift(1).rolling(window=50, min_periods=1).mean()*0.3) + (a['PSxG'].shift(1).rolling(window=30, min_periods=1).mean()*0.1)), 2)

    new = a[["date", "time", "team", "goals_offense", "goals_defense"]]

    b = data[data.columns[8:]]
    c = data[data.columns[:8]]

    for i in b.columns:
        b[i] = round(b[i] / ((b[i].shift(1).rolling(window=100, min_periods=1).quantile(0.9)*0.6) + (b[i].shift(1).rolling(window=50, min_periods=1).quantile(0.9)*0.3) + (b[i].shift(1).rolling(window=30, min_periods=1).quantile(0.9)*0.1)), 2)

    all = pd.concat([c, b], axis=1)
    all = all.fillna(0)

    all['yellows'] = all['yellows'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['reds'] = all['reds'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['failed_controls'] = all['failed_controls'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['dispossession_after_tackle'] = all['dispossession_after_tackle'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['double_yellows'] = all['double_yellows'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['fouls'] = all['fouls'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['unsuccessful_dribbler_challenge'] = all['unsuccessful_dribbler_challenge'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['error_leading_to_opponent_shot'] = all['error_leading_to_opponent_shot'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['player_take_ons_tackled'] = all['player_take_ons_tackled'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['penalty_kicks_conceded'] = all['penalty_kicks_conceded'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['own_goals'] = all['own_goals'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))
    all['aerials_lost'] = all['aerials_lost'].apply(lambda x: (1 - x) if pd.notnull(x) else x).apply(lambda x: max(0, x))

    all["offense"] = all[["assists", "shots", "shots_on_target", "PSxG", "non_penalty_expected_goals", "expected_assisted_goals", "shot_creating_actions", "goal_creating_actions", "progressive_passes", "progressive_carries", "progressive_passing_distance", "expected_assists", "key_passes", "passes_into_final_third", "passes_into_penalty_area", "crosses_into_penalty_area", "through_balls", "crosses", "corner_kicks", "passes_offside", "touches_in_attacking_penalty_area", "carries_into_penalty_box", "offsides", "penalty_kicks_won"]].mean(axis=1).round(2)
    all["in_game"] = all[["yellows", "reds", "touches", "passes_completed", "carries", "failed_controls", "dispossession_after_tackle", "double_yellows", "fouls"]].mean(axis=1).round(2)
    all["defense"] = all[["tackles", "interceptions", "blocks_by_player", "tackles_won", "dribbler_tackles_attempts", "dribblers_tackled", "unsuccessful_dribbler_challenge", "clearances", "error_leading_to_opponent_shot", "player_take_ons_tackled", "recoveries", "aerials_won", "aerials_lost"]].mean(axis=1).round(2)

    all = all[["team", "score", "score_against", "xG", "xg_against", "offense", "in_game", "defense"]]

    all = all[all["team"]==team]

    all["offense"] = round(((all['offense'].shift(1).rolling(window=100, min_periods=1).mean()*0.5) + (all['offense'].shift(1).rolling(window=50, min_periods=1).mean()*0.3) + (all['offense'].shift(1).rolling(window=30, min_periods=1).mean()*0.2))*(1 + (all["score"].shift(1) * 0.05) + (all["xG"].shift(1) * 0.05)), 2)
    all["in_game"] = round(((all['in_game'].shift(1).rolling(window=100, min_periods=1).mean()*0.5) + (all['in_game'].shift(1).rolling(window=50, min_periods=1).mean()*0.3) + (all['in_game'].shift(1).rolling(window=30, min_periods=1).mean()*0.2)), 2)
    all["defense"] = round(((all['defense'].shift(1).rolling(window=100, min_periods=1).mean()*0.5) + (all['defense'].shift(1).rolling(window=50, min_periods=1).mean()*0.3) + (all['defense'].shift(1).rolling(window=30, min_periods=1).mean()*0.2))*(1 - (all["score_against"].shift(1) * 0.05) - (all["xg_against"].shift(1) * 0.05)), 2)

    all.drop(["team", "xG", "xg_against"], axis=1, inplace=True)
    
    complete = pd.concat([new, all], axis=1)
    return complete
```

```python
aa = pd.read_csv("../tables/fbref.com/17-18.csv", index_col=[0]).reset_index().drop("index", axis=1)
bb = pd.read_csv("../tables/fbref.com/18-19.csv", index_col=[0]).reset_index().drop("index", axis=1)
cc = pd.read_csv("../tables/fbref.com/19-20.csv", index_col=[0]).reset_index().drop("index", axis=1)
dd = pd.read_csv("../tables/fbref.com/20-21.csv", index_col=[0]).reset_index().drop("index", axis=1)
ee = pd.read_csv("../tables/fbref.com/21-22.csv", index_col=[0]).reset_index().drop("index", axis=1)
ff = pd.read_csv("../tables/fbref.com/22-23.csv", index_col=[0]).reset_index().drop("index", axis=1)
gg = pd.read_csv("../tables/fbref.com/23-24.csv", index_col=[0]).reset_index().drop("index", axis=1)

data = pd.concat([aa, bb, cc, dd, ee, ff, gg])

teams = data["team"].unique()

concatenate = []

for i in teams:
    if len(concatenate) == 0:
        concatenate = ratings(i)
    else:
        concatenate = pd.concat([concatenate, ratings(i)])
```

```python
concatenate = concatenate
concatenate = concatenate.sort_values(by=["date", "time"]).reset_index().drop(["index"], axis=1)
concatenate.groupby("team")["offense"].mean().sort_values(ascending=False)
```

```python
concatenate.drop(["score_against"], axis=1, inplace=True)
concatenate = concatenate.reset_index().drop(["index"], axis=1)
even_rows = concatenate.iloc[::2]  # Selecting even index rows
odd_rows = concatenate.iloc[1::2]  # Selecting odd index rows

# Duplicate and modify the index for odd rows
odd_rows.index = even_rows.index

# Concatenate odd rows to the bottom of even rows
result_df = pd.concat([even_rows.rename(columns=lambda x: f'home_{x}'), odd_rows.rename(columns=lambda x: f'away_{x}')], axis=1)

result_df.drop(["away_date", "away_time"], axis=1, inplace=True)
result_df.rename(columns={'home_date': 'date', 'home_time': 'time'}, inplace=True)
result_df['result'] = np.where(result_df['home_score'] > result_df['away_score'], 1, np.where(result_df['home_score'] == result_df['away_score'], 0.5, 0))

result_df.sort_values(by=["date", "time"], inplace=True)
```

```python
result_df = result_df.reset_index().drop(["index"], axis=1)
result_df.drop(["home_score", "away_score"], axis=1, inplace=True)
result_df['date'] = pd.to_datetime(result_df['date'])
```

```python
def home_rank(row):
    target_date = row['date']
    target_value = row['home_team']

    # Filter DataFrame 1 where 'Value' equals the target value and the date is strictly less than the target date
    filtered_df = season[(season['team'] == target_value) & (pd.to_datetime(season['date']) < pd.to_datetime(target_date))]

    if not filtered_df.empty:
        # Find the index of the maximum date within the filtered DataFrame
        closest_date_index = filtered_df['date'].idxmax()

        # Get the value in the "Rank" column for the row with the closest date and 'Value' equals the target value
        result_rank = season.loc[closest_date_index, 'rank']
        return result_rank
    else:
        # Set a default value (e.g., None or -1) when there is no date strictly less than the target date
        return 0

# Apply the function to DataFrame 2 and create a new column 'Closest_Rank'
result_df['home_standing'] = result_df.apply(home_rank, axis=1)

def away_rank(row):
    target_date = row['date']
    target_value = row['away_team']

    # Filter DataFrame 1 where 'Value' equals the target value and the date is strictly less than the target date
    filtered_df = season[(season['team'] == target_value) & (pd.to_datetime(season['date']) < pd.to_datetime(target_date))]

    if not filtered_df.empty:
        # Find the index of the maximum date within the filtered DataFrame
        closest_date_index = filtered_df['date'].idxmax()

        # Get the value in the "Rank" column for the row with the closest date and 'Value' equals the target value
        result_rank = season.loc[closest_date_index, 'rank']
        return result_rank
    else:
        # Set a default value (e.g., None or -1) when there is no date strictly less than the target date
        return 0

# Apply the function to DataFrame 2 and create a new column 'Closest_Rank'
result_df['away_standing'] = result_df.apply(away_rank, axis=1)
```
